```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(y,x)
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
```

```python
fig = plt.figure()

axes = fig.add_axes([.1,.1,.8,.8])

axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Big Plot')

axes2.plot(y,x,'r')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Small Plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current in axes:      #iteration
    current.plot(x,y) 
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

axes[0].plot(x,y)
axes[0].set_title('x,y')

axes[1].plot(y,x)
axes[1].set_title('y,x')

axes[2].plot(x+y,x-y)
axes[2].set_title('x+y,x-y')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(4,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='x**2')
axes.plot(x,x**3,label='x**3')

axes.legend(loc=1) # locations in documentation (0 best). works with coordinates in tuples too.
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, color='#1FC241', linewidth = 2, alpha = 1, linestyle = '--', 
          marker = 'o', markersize = 10, markerfacecolor = 'blue', markeredgewidth=2, markeredgecolor = 'red')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.set_xlim([2,4])
axes.set_ylim([5,11])
axes.plot(x,y)
```
